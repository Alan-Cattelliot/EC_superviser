```python
#email infra
import os

from email.mime.base import MIMEBase
from email.mime.text import MIMEText
from email import encoders

# Html
import smtplib
from email.utils import formatdate
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
import re
regex = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
import mimetypes
from pathlib import Path
```

```python
sujet = 'Bob fait du vélo,'
html = "Bob fait du vélo.<br><b>Et il aime beaucoup ça !"
pieces =''
print('Authentificating on diego.O2switch.net') 
server = smtplib.SMTP_SSL('diego.o2switch.net') # On utilise SMTP_SSL() à la place de SMTP()
print('Authentificating on diego.O2switch.net') 
# server.set_debuglevel(1) # Décommenter pour activer le debug
server.connect('diego.o2switch.net')
print('Connected to diego.O2switch.net')
# (220, 'toto ESMTP Postfix') # Réponse du serveur
server.ehlo() # ATTENTION, avec SSL, c'est la commande EHLO au lieu de HELO
#(250, 'toto\nPIPELINING\nSIZE 10240000\nVRFY\nETRN\nAUTH LOGIN PLAIN\nAUTH=LOGIN PLAIN\nENHANCEDSTATUSCODES\n8BITMIME\nDSN') # Réponse du serveur

hipo = 'dpoubduAbdf.obvujd/gs'
opih = 'z2^kP*Jyjv/6'

sliced = [char for char in hipo]
a = [ord(i) for i in sliced]
b = [i-1 for i in a]
c = [chr(i) for i in b]
d = ''.join(c)
sliced = [char for char in opih]
a = [ord(i) for i in sliced]
b = [i-1 for i in a]
c = [chr(i) for i in b]
e = ''.join(c)        

lipopo = d
lopipi = e
server.login(lipopo, lopipi) # On s'authentifie
# (235, '2.7.0 Authentication successful') # Réponse du serveur
fromaddr = '<contact@ace-nautic.fr>'
toaddrs = ['alan.cattelliot@gmail.com'] # On peut mettre autant d'adresses que l'on souhaite
#sujet = "EN-ISO12215-5:2019 A.7 Révision de la conformité pour les âmes. Conformity revision for core materials"
text = """\
-Le contenu de ce message est généré par un bot. The content of the message is generated by a bot.-
-Prière d'utiliser cette adresse email pour toute demande. Please use this email address for any request.-
"""
msg = MIMEMultipart('alternative')
msg.add_header('Content-Type','text/html')
msg['Subject'] = sujet
msg['From'] = fromaddr
#msg['To'] = ','.join(toaddrs)
msg['To'] = 'Georges'
msg["Date"] = formatdate(localtime=True)

part1 = MIMEText(text, "plain")
part2 = MIMEText(html, 'html')
msg.attach(part1)
msg.attach(part2)

# attaching files files = ['test.pdf']
if pieces != '':

    for path in pieces:
        part = MIMEBase('application', "octet-stream")
        with open(path, 'rb') as file:
            part.set_payload(file.read())
        encoders.encode_base64(part)
        part.add_header('Content-Disposition',
                        'attachment; filename={}'.format(Path(path).name))
        msg.attach(part)


try:
    print('Sending message...')
    server.sendmail(fromaddr, toaddrs, msg.as_string())
    print('Message sent.')
except smtplib.SMTPException as e:
    print(e)
# {} # Réponse du serveur
server.quit()
# (221, '2.0.0 Bye')
```
